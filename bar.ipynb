### 所有性能分5组统计的柱状图（二分类）

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取 Excel 文件
df = pd.read_excel(r'C:\Users\ynn\Desktop\result\binary.xlsx')

# 将数据转换为字典列表
data = df.to_dict('records')

metrics = ['acc', 'f1', 'precision', 'recall']
group_values = sorted(set(d['group'] for d in data))
x = np.arange(len(group_values))
width = 0.15

fig, ax = plt.subplots(figsize=(12, 6))

# 手动定义颜色
colors = ['#03045E', '#0077B6', '#00D4B8', '#90E0EF']

for i, metric in enumerate(metrics):
    ax.bar(x + i * width, [next(d[metric] for d in data if d['group'] == gv) for gv in group_values], 
           width, label=metric, color=colors[i])

ax.set_xticks(x + width * (len(metrics) - 1) / 2)
ax.set_xticklabels([str(gv) for gv in group_values])
ax.set_xlabel('Group')
ax.set_ylabel('Value')
ax.set_title('Comparison of Metrics across Groups')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(False)
# ax.legend().set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.subplots_adjust(right=0.8)
plt.show()
```
